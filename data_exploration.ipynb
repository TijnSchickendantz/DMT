```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 200)
```

```python
df = pd.read_csv('dataset_mood_smartphone.csv')
df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d %H:%M:%S.%f')
df['date'] = df['time'].dt.date
```

```python
fig = plt.figure(figsize=(4,2.5), dpi=300)
df = pd.read_csv('dataset_mood_smartphone.csv')
moods = df[df['variable']=='mood']['value']
sns.histplot(moods, bins=10)
plt.xticks(np.arange(1,11))
plt.xlabel('Mood')
plt.savefig('plots/mood_hist.png', bbox_inches='tight')
plt.show()
```

```python
variables = df['variable'].unique()
variable = 'appCat.communication'
df['ID_num'] = df['id'].apply(lambda x: x[-2:])

fig = plt.figure(figsize=(3,2), dpi=300)
plt.scatter(df[df['variable']==variable]['ID_num'], df[df['variable']==variable]['value'], s=6)
plt.xlabel('ID')
plt.ylabel(variable)
plt.xticks(['01', '17', '33'])
plt.savefig('plots/outliers.png', bbox_inches='tight')
plt.show()
```

```python
id = 'AS14.01'
df_id = df[(df['id']==id) & (df['variable']=='mood')][['time', 'value']]

df_id.plot(x='time', y='value', color='k', figsize=(4,2.5))
plt.legend([])
plt.ylabel('Mood')
plt.xlabel('Date')
plt.title(id)
plt.gcf().autofmt_xdate()
plt.savefig('plots/mood_example.png', bbox_inches='tight')
plt.show()
```

```python
variables = df['variable'].unique()

for variable in variables:

    # scatterplot
    plt.scatter(df[df['variable']==variable]['id'], df[df['variable']==variable]['value'], s=0.5)
    plt.xlabel('ID')
    plt.ylabel(variable)
    plt.show()
```
